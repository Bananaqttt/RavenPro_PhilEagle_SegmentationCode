```python
import csv
import os

# ---------------------------------------------------------
# CONFIGURATION
# ---------------------------------------------------------
# Replace this path with the folder containing your audio files.
# Example (Windows): r"C:\Users\Name\Documents\AudioData"
# Example (Mac/Linux): "/Users/name/documents/audiodata"
base_directory = r"YOUR_SPECIFIED_FOLDER_PATH_HERE"

# The list of files to remove (generated by me)
delete_list_file = 'files_to_delete.csv'
# ---------------------------------------------------------

def delete_files():
    if not os.path.exists(delete_list_file):
        print(f"Error: Could not find {delete_list_file} in the current directory.")
        return

    print(f"Reading {delete_list_file}...")
    
    deleted_count = 0
    not_found_count = 0

    with open(delete_list_file, mode='r', newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        
        for row in reader:
            # Construct the full file path
            # The manifest contains backslashes in 'output_folder'.
            # We replace them with the correct separator for your OS.
            relative_folder = row['output_folder'].replace('\\', os.sep)
            filename = row['segment_filename']
            
            # Full path construction
            file_path = os.path.join(base_directory, relative_folder, filename)
            
            try:
                if os.path.exists(file_path):
                    os.remove(file_path)
                    print(f"Deleted: {filename}")
                    deleted_count += 1
                else:
                    # Fallback: Check if file is directly in base_directory (flat structure)
                    flat_path = os.path.join(base_directory, filename)
                    if os.path.exists(flat_path):
                        os.remove(flat_path)
                        print(f"Deleted (flat): {filename}")
                        deleted_count += 1
                    else:
                        print(f"Not Found: {file_path}")
                        not_found_count += 1
                        
            except Exception as e:
                print(f"Error deleting {filename}: {e}")

    print("-" * 30)
    print(f"Process Complete.")
    print(f"Total files deleted: {deleted_count}")
    print(f"Files not found: {not_found_count}")

if __name__ == "__main__":
    delete_files()
```
